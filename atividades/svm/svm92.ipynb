### Importar as bibliotecas

```python
import pandas as pd
# import seaborn as sns
# import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn import svm
```

### Ler dados

```python
df = pd.read_csv('../../sources/online_shoppers_intention.csv')

df
```

### Tratar dados

```python
df.dtypes
```

```python
df.isna().sum()
```

Tem uma quantidade negligenciavel de nulos, so remover

```python
df = df.dropna()
```

conversão do mês

```python
le = LabelEncoder()

df['Month'] = le.fit_transform(df['Month'])
df['VisitorType'] = le.fit_transform(df['VisitorType'])

df
```

```python
df['Revenue'].value_counts()
```

Balanciando os dados para não enviesar o SVM

```python
from imblearn.over_sampling import SMOTE

vars_indep = df.iloc[:,0:17]
var_dep = df.iloc[:,17]

vars_indep_res, var_dep_res = SMOTE(random_state=777).fit_resample(vars_indep, var_dep)
```

### Treinamento do modelo

```python
sc = StandardScaler()

vars_indep_scaler = sc.fit_transform(vars_indep_res)

vars_indep_treino, vars_indep_teste, var_dep_treino, var_dep_teste = train_test_split(vars_indep_scaler, var_dep_res, test_size=0.7, random_state=777)

model = svm.SVC(kernel='linear')

model.fit(vars_indep_treino, var_dep_treino)
```

### Medindo o modelo

Com dados de treino

```python
var_dep_treino_prev = model.predict(vars_indep_treino)

print(accuracy_score(var_dep_treino, var_dep_treino_prev))
print(confusion_matrix(var_dep_treino, var_dep_treino_prev))
```

Com dados de teste

```python
var_dep_teste_prev = model.predict(vars_indep_teste)

print(accuracy_score(var_dep_teste, var_dep_teste_prev))
print(confusion_matrix(var_dep_teste, var_dep_teste_prev))
```

Melhorando a acuracidade

```python
from sklearn.svm import SVC

kernels = ['rbf']
c_range = [650, 800]
gamma_range = [0.0090, 0.0095, 0.0099]

svm_param_grid = {
    'gamma': gamma_range, 
    'C': c_range,
    'kernel': kernels
}

grid_search = GridSearchCV(SVC(), param_grid=svm_param_grid, cv=3)

model = grid_search.fit(vars_indep_treino, var_dep_treino)

var_dep_treino_prev = model.predict(vars_indep_treino)

var_dep_teste_prev = model.predict(vars_indep_teste)

print('====================================================')
print('BEST SCORE', model.best_score_)
print('BEST PARAMS', model.best_params_)
print('====================================================')
print('TRAIN ACCURACY', accuracy_score(var_dep_treino, var_dep_treino_prev))
print('TRAIN CONFUSION MATRIX', confusion_matrix(var_dep_treino, var_dep_treino_prev))
print('====================================================')
print('TEST ACCURACY', accuracy_score(var_dep_teste, var_dep_teste_prev))
print('TEST CONFUSION MATRIX', confusion_matrix(var_dep_teste, var_dep_teste_prev))
```
