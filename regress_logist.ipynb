### Ler os dados

```python
import pandas as pd

df = pd.read_csv('./sources/source1.csv')

df
```

### Lidar com nulos

```python
df.apply(lambda x: sum(x.isnull()), axis=0)
```

```python
df['Gender'].fillna(df['Gender'].mode()[0], inplace=True)
df['Married'].fillna(df['Married'].mode()[0], inplace=True)
df['Dependents'].fillna(df['Dependents'].mode()[0], inplace=True)
df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mode()[0], inplace=True)
df['Credit_History'].fillna(df['Credit_History'].mode()[0], inplace=True)
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)
df['Self_Employed'].fillna('No', inplace=True)

df.apply(lambda x : sum(x.isnull()), axis=0)
```

### Transformar texto em número

```python
from sklearn.preprocessing import LabelEncoder

df = df.drop(axis=1, columns=['Loan_ID'])

le = LabelEncoder()

for i in df.columns:
    df[i] = le.fit_transform(df[i])

df
```

### Separar base de treino e teste

```python
from sklearn.model_selection import train_test_split

input_dados = df.drop(axis=1, columns=['Loan_Status'])

output_dados = df[['Loan_Status']]

# x_treino, x_teste, y_treino, y_teste
input_treino, input_teste, output_treino, output_teste = train_test_split(
    input_dados, output_dados,  test_size=0.3, random_state=777)
```

### Treinar o modelo

```python
from sklearn.linear_model import LogisticRegression

# model = LogisticRegression().fit(input_treino, output_treino) -> computacionalmente intensivo treinar
model = LogisticRegression(max_iter=1000).fit(input_treino, output_treino) # -> limita a quantidade de vezes que tenta corrigir a função do modelo

output_prev_treino = model.predict(input_treino)

output_prev_treino
```

### Medindo o treino

```python
from sklearn.metrics import accuracy_score, confusion_matrix

print("####### TREINO")
print(accuracy_score(output_treino, output_prev_treino)) # percentual de acerto
print(confusion_matrix(output_treino, output_prev_treino)) # 

output_prev_teste = model.predict(input_teste)

print("####### TESTE")
print(accuracy_score(output_teste, output_prev_teste))
print(confusion_matrix(output_teste, output_prev_teste))
```
